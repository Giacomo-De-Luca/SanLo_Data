```python
##generate users belonging to a particular age group
import random
import string
import json
import csv

# Function to generate a random string (for user and device names)
def random_string(length):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for i in range(length))

# Function to generate random user data with specified age groups
def generate_random_user_data(num_users):
    age_groups = [(0, 12), (12, 18), (18, 28), (28, 40), (40, 50), (60, 90)]
    weights = [5, 10, 20, 15, 10, 5]  # Weights for each age group

    users = []
    for user_id in range(1, num_users + 1):
        age_group = random.choices(age_groups, weights=weights)[0]
        age = f"{age_group[0]}-{age_group[1]}"

        user = {
            "id": user_id,
            "nickname": f"user_{user_id}",
            "device_id": f"device_{user_id}",
            "age": age,
            "registration_time": (datetime.utcnow() - timedelta(days=random.randint(1, 365))).isoformat()
        }
        users.append(user)

    return users

# Function to save user data in JSON format
def save_user_data_json(users, filename):
    with open(filename, "w") as file:
        json.dump(users, file, indent=4)

# Function to save user data in CSV format
def save_user_data_csv(users, filename):
    with open(filename, "w", newline="") as file:
        fieldnames = ["id", "nickname", "device_id", "age", "registration_time"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for user in users:
            writer.writerow(user)

# Example usage to generate and save 100 user data in both JSON and CSV formats
if __name__ == "__main__":
    num_users_to_generate = 100
    users = generate_random_user_data(num_users_to_generate)

    # Save user data in JSON
    json_filename = "user_data3.json"
    save_user_data_json(users, json_filename)

    # Save user data in CSV
    csv_filename = "user_data3.csv"
    save_user_data_csv(users, csv_filename)
```

```python
## generate tower data and enemy data // al momento non lo sto usando, i dati delle torri sono già nei game data 

import random
import string
import json
import csv

# Function to generate a random string (for tower and enemy names)
def random_string(length):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for i in range(length))

# Function to generate random tower data
def generate_random_tower_data(num_towers):
    towers = []

    for tower_id in range(1, num_towers + 1):
        tower = {
            "id": tower_id,
            "tower_name": f"tower_{tower_id}",
            "avg_damage": random.randint(5, 50),
            "count": random.randint(0, 100)
        }
        towers.append(tower)

    return towers

# Function to generate random enemy data
def generate_random_enemy_data(num_enemies):
    enemies = []

    for enemy_id in range(1, num_enemies + 1):
        enemy = {
            "id": enemy_id,
            "name": f"enemy_{enemy_id}",
            "avg_damage": random.randint(1, 10),
            "count": random.randint(0, 50)
        }
        enemies.append(enemy)

    return enemies

# Function to save tower data in JSON format
def save_tower_data_json(towers, filename):
    with open(filename, "w") as file:
        json.dump(towers, file, indent=4)

# Function to save tower data in CSV format
def save_tower_data_csv(towers, filename):
    with open(filename, "w", newline="") as file:
        fieldnames = ["id", "tower_name", "avg_damage", "count"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for tower in towers:
            writer.writerow(tower)

# Function to save enemy data in JSON format
def save_enemy_data_json(enemies, filename):
    with open(filename, "w") as file:
        json.dump(enemies, file, indent=4)

# Function to save enemy data in CSV format
def save_enemy_data_csv(enemies, filename):
    with open(filename, "w", newline="") as file:
        fieldnames = ["id", "name", "avg_damage", "count"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for enemy in enemies:
            writer.writerow(enemy)

# Example usage to generate and save Tower data in both JSON and CSV formats
if __name__ == "__main__":
    num_towers_to_generate = 5
    towers = generate_random_tower_data(num_towers_to_generate)

    # Save Tower data in JSON
    json_filename = "tower_data.json"
    save_tower_data_json(towers, json_filename)

    # Save Tower data in CSV
    csv_filename = "tower_data.csv"
    save_tower_data_csv(towers, csv_filename)

    # Example usage to generate and save Enemy data in both JSON and CSV formats
    num_enemies_to_generate = 5
    enemies = generate_random_enemy_data(num_enemies_to_generate)

    # Save Enemy data in JSON
    json_filename = "enemy_data.json"
    save_enemy_data_json(enemies, json_filename)

    # Save Enemy data in CSV
    csv_filename = "enemy_data.csv"
    save_enemy_data_csv(enemies, csv_filename)
```
